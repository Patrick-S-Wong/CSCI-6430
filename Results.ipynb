```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in data from file
dtResults = np.loadtxt("decision_tree_results.txt")
rfResults = np.loadtxt("random_forest_results.txt")
mlpResults = np.loadtxt("ML_perceptron_results.txt")

# Generate box and whiskers plot from list of results
plt.boxplot([dtResults, rfResults, mlpResults])

# Create labels and titles
plt.title("Breast cancer classification testing accuracy")
plt.xlabel("Model")
plt.ylabel("Accuracy")

# add x-axis tick labels
plt.xticks([1, 2, 3], ['Decision Tree', 'Random Forest', 'ML Perceptron'])

# show plot
plt.show()
```

```python
results = [dtResults, rfResults, mlpResults]
```

```python
type(results[0])
```

```python
for item in range(len(results)):
    data = results[item]
    
    median = np.median(data)
    upper_quartile = np.percentile(data, 75)
    lower_quartile = np.percentile(data, 25)

    iqr = upper_quartile - lower_quartile
    upper_whisker = data[data<=upper_quartile+1.5*iqr].max()
    lower_whisker = data[data>=lower_quartile-1.5*iqr].min()
    
    print("Data for results", item)
    print("Median", '{number:.{digits}f}'.format(number=median, digits=4))
    print("Upper quartile", '{number:.{digits}f}'.format(number=upper_quartile, digits=4))
    print("Lower quartile", '{number:.{digits}f}'.format(number=lower_quartile, digits=4))
    print("Inter quartile range", '{number:.{digits}f}'.format(number=iqr, digits=4))
    print("Upper whisker", '{number:.{digits}f}'.format(number=upper_whisker, digits=4))
    print("Lower whisker", '{number:.{digits}f}'.format(number=lower_whisker, digits=4))
    print()
```
